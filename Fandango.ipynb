```python
# FANDANGO VS ALL SITES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Fandango=pd.read_csv(r'C:\Users\LENOVO\Documents\VIT\CODES\AI ML\CAPSTONE\fandango_scrape.csv')
all_sites=pd.read_csv(r'C:\Users\LENOVO\Documents\VIT\CODES\AI ML\CAPSTONE\all_sites_scores.csv')
```

```python
Fandango.head()
```

```python
all_sites.head()
```

```python
Merged=pd.merge(left=Fandango,right=all_sites,how='inner',on='FILM')
Merged.head()
```

```python
# normalizing all the ratings for all sites
Merged['RottenTomatoes_NORM']=Merged['RottenTomatoes']/20
Merged['RottenTomatoes_User_NORM']=Merged['RottenTomatoes_User']/20
Merged['Metacritic_NORM']=Merged['Metacritic']/20
Merged['Metacritic_User_NORM']=Merged['Metacritic_User']/2
Merged['IMDB_NORM']=Merged['IMDB']/2
Merged.head()
```

```python
Merged.columns
```

```python
df=Merged[['STARS', 'RATING','RottenTomatoes_NORM', 'RottenTomatoes_User_NORM', 'Metacritic_NORM',
       'Metacritic_User_NORM', 'IMDB_NORM']]
df.head()
```

```python
plt.figure(figsize=(10,7),dpi=150)
sns.kdeplot(data=df,x='RATING',fill=True,label='Fandango rating',clip=(0,5))
sns.kdeplot(data=df,x='STARS',fill=True,label='Fandango Stars',clip=(0,5))
sns.kdeplot(data=df,x='RottenTomatoes_NORM',fill=True,label='RottenTomatoes_NORM',clip=(0,5))
sns.kdeplot(data=df,x='RottenTomatoes_User_NORM',fill=True,label='RottenTomatoes_User_NORM',clip=(0,5))
sns.kdeplot(data=df,x='Metacritic_NORM',fill=True,label='Metacritic_NORM',clip=(0,5))
sns.kdeplot(data=df,x='Metacritic_User_NORM',fill=True,label='Metacritic_User_NORM',clip=(0,5))
sns.kdeplot(data=df,x='IMDB_NORM',fill=True,label='IMDB_NORM',clip=(0,5))
plt.legend(loc=(1.05,0.5))
```

```python
# From the plot its clear that Fandango has an uneven distribution whereas other sites have fairly uniformly distributed values
```

```python
norm_movies=df[['FILM','STARS', 'RATING','RottenTomatoes_NORM', 'RottenTomatoes_User_NORM', 'Metacritic_NORM',
       'Metacritic_User_NORM', 'IMDB_NORM']]
norm_movies.head()
```

```python
# lowest rated movies acc to rotten tomatoes critic rating
worst_films=norm_movies.nsmallest(10,'RottenTomatoes_NORM')
```

```python
# visualising data of 10 worst movies
sns.kdeplot(data=worst_films,x='STARS',clip=(0,5),fill=True,label='STARS')
sns.kdeplot(data=worst_films,x='RATING',clip=(0,5),fill=True,label='RATING')
sns.kdeplot(data=worst_films,x='RottenTomatoes_NORM',clip=(0,5),fill=True,label='RottenTomatoes_NORM')
sns.kdeplot(data=worst_films,x='RottenTomatoes_User_NORM',clip=(0,5),fill=True,label='RottenTomatoes_User_NORM')
sns.kdeplot(data=worst_films,x='Metacritic_NORM',clip=(0,5),fill=True,label='Metacritic_NORM')
sns.kdeplot(data=worst_films,x='Metacritic_User_NORM',clip=(0,5),fill=True,label='Metacritic_User_NORM')
sns.kdeplot(data=worst_films,x='IMDB_NORM',clip=(0,5),fill=True,label='IMDB_NORM')
plt.legend(loc=(1.05,0.5))
```
